```python
# №1 — корень n-й степени из числа x

def num_sqrt(x, n):
    return round(x ** (1/n), 4) 

x, n = int(input()), int(input())
print(num_sqrt(x, n))
```

```python
# №2 — функция которая принимает список расширений и имя файла

def ext_check(file, extensions):
    if file[file.rfind('.') + 1:] in extensions:
        return "Файл имеет правильное расширение"
    else:
        return "У файла неверное расширение"
    
file = 'Моя диссертация.svg'
extensions = ['png', 'jpg', 'jpeg', 'gif', 'svg']

print(ext_check(file, extensions))
```

```python
# №4 — складывать цифры внутри числа

def digit_sum(n):
    digit = 0
    for i in range(len(str(n))):
        digit += n % 10
        n = int(n / 10)
    return digit

n = int(input())
print(digit_sum(n))
```

```python
# №5 — числа в последовательности уникальны

numbers=[1,4,5,6,7,0]

num_set = set(numbers)
num_set = list(num_set)
numbers.sort()

if num_set == numbers:
    print("Все числа в последовательности уникальны!!!")
else:
    print("Числа не уникальны")
```

```python
# №6 — печатает все элементы заданного ей списка

def print_list(my_list):
    [print(f"НОМЕР {i} --> {my_list[i]}") for i in range(len(my_list))]

print_list(["я", "не", "в", "отпуске", "и", "это", "грустно"])
```

```python
# №7 — печатает каждый элемент словарика

def print_dict(my_dict):
    [print("Ключ", key, "-->", value) for key, value in my_dict.items()]
        
print_dict({"key1": 2, "key3": False, "Приветствие": "Hello"})
```

```python
# №8 — печатает каждый элемент словарика или списка

def print_overlord(my_dict):
    for i in my_dict.values():
        if type(i) == dict:
            [print("Ключ", key, "-->", value) for key, value in i.items()]
        elif type(i) == list:
            [print(f"НОМЕР {j} --> {i[j]}") for j in range(len(my_dict))]
        else:
            print(i)

print_overlord(dict(key1=1,
            key2=[1, 2, 3, 4], 
            key3='Hello', 
            key4={"ciao":"Mondo", "Привет": "О дивный мир"}))
```

```python
# №9 — токенизация

def word_count(phrase):
    phrase = phrase.split(" ")
    for i in range(len(phrase)):
        print(f"НОМЕР {i} --> {phrase[i]}")
    
word_count("Тут фраза с пробелами и я хочу их распечатать по одному на строку и ещё посчитать слова")
```

```python
# №10 — "вес" каждого слова

def word_count(phrase):
    symbols = [",", ".", "!", "?", ":", ";"]
    phrase = phrase.split(" ")
    weight = list()
    
    for i in range(len(phrase)):    # удаляет символы ",", ".", "!", "?", ":", ";" после слова
        if phrase[i][-1:] in symbols:
            phrase[i] = phrase[i][:-1]
            
    for i in range(len(phrase)):
        counter = 0
        for j in range(len(phrase)):
            if phrase[i].lower() == phrase[j].lower():
                counter += 1
        weight.append((counter/len(phrase)) * 100) 
        
    for i in range(len(phrase)):
        print(f"НОМЕР {i} --> {phrase[i]} ({round(weight[i], 2)}%) ")
    
word_count("Тут фраза с пробелами и я хочу их распечатать по одному на строку и ещё посчитать слова")
print('============')
word_count("Подсказка: тут уже не подсказка получится просто создать переменную от 0 до чего-то и перебирать по ней просто увеличивая её на один, потому что в словарике ключи несортированы и непоследовательны. Это, кстати, означает заодно то что порядок вывода в функции может отличаться от того, в каком вы создавали словарик. Тут вам понадобится либо метод keys либо метод items, и может быть, такие понятия как \"распаковка\". Цикл будет сделан с помощью оператора for. Эту задачку можно решить и без цикла через list comprehension. Про работу со словариками посмотрите конец занятия L1.1")
```
